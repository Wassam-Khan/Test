```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```python
X = np.array([[20], [22], [24], [26], [28], [30], [32], [34]]).astype(float)
y = np.array([180, 195, 210, 230, 255, 270, 300, 320]).astype(float)
```

```python
scaler_X = StandardScaler()
scaler_y = StandardScaler()
```

```python
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).ravel()
```

```python
svr = SVR(kernel='rbf', C=100, epsilon=0.1)
svr.fit(X_scaled, y_scaled)
```

```python
y_pred_scaled = svr.predict(X_scaled)
```

```python
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1)).ravel()
```

```python
#Now we plot the actual vs predicted
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='green', label='Actual Ice Cream Sales')
plt.plot(X, y_pred, color='yellow', label='SVR Prediction')
plt.title("SVR - Ice Cream Sales vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Creams Sold")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
